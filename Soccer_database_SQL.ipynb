Find all matches from a specific league, need to join on leagueID

```python
# Select all from Leauge
League = pd.read_sql("""
                   SELECT *
                   FROM League
                   """, conn)

League
```

```python
# Select all from Match
Match = pd.read_sql("""
                   SELECT *
                   FROM match
                   """, conn)

pd.set_option('display.max_columns', None)
Match
```

```python
# Inner join Team and Match, select teams where the leage was Italy Series A. 

Match_League = pd.read_sql("""
                   SELECT m.match_api_id AS matchID, home_team_api_id AS HomeTeam, away_team_api_id AS AwayTeam, l.name as League
                   FROM Match as m
                   INNER JOIN League as l
                   ON l.id = m.league_id
                   WHERE League = 'Italy Serie A'
                   """, conn)

Match_League
```

```python
# Select match details from the Match table where the league ID matches a specific value. You can join this with the League table to get the league name.

Match_League = pd.read_sql("""
                   SELECT m.*, l.name 
                   FROM Match as m
                   INNER JOIN League as l
                   ON l.id = m.league_id
                   WHERE league_id = '1'
                   """, conn)

Match_League
```

```python
#  Calculate the sum of home and away goals for each match. Show total goals scored in a match.

Match_goals = pd.read_sql("""
                   SELECT home_team_goal + away_team_goal AS total_goals
                   FROM Match
                   """, conn)


# Displaying the DataFrame
print(Match_goals)

Match_goals
```

```python
# Find the top 10 players with the highest overall rating.
# Order Descending and limit to 10

# Select all from players
Players = pd.read_sql("""
                   SELECT player_api_id, overall_rating
                   FROM player_attributes
                   ORDER BY overall_rating DESC
                   limit 10;
                   """, conn)

pd.set_option('display.max_columns', None)
Players
```

```python
# Select player names and their ratings from the Player_Attributes table, joining with the Player table, and order by rating descending.

# Select all from players
Players_rating = pd.read_sql("""
                   SELECT p.player_name, pa.overall_rating
                   FROM Player AS p
                   INNER JOIN player_attributes AS pa
                   ON p.player_api_id = pa.player_api_id
                   ORDER BY overall_rating DESC;
                   """, conn)

pd.set_option('display.max_columns', None)
Players_rating
```

```python
# Determine the average rating of players by position.
# player attributes table; positioning and overall_rating
# what is the average rating for all players in X position

# Select all from players
Players_positioning = pd.read_sql("""
                   SELECT positioning, AVG(overall_rating)
                   FROM Player_attributes
                   GROUP BY positioning;
                
                   """, conn)

pd.set_option('display.max_columns', None)
Players_positioning
```

```python
# List matches with more than 3 goals scored in total.

# Select all from players
Match_goals = pd.read_sql("""
                   SELECT home_team_goal + away_team_goal AS overall_goal
                   FROM Match
                   WHERE overall_goal > 3;
                   """, conn)

pd.set_option('display.max_columns', None)
Match_goals
```
